```julia
# Rabbits and Foxes:
# An example of the classical predator/prey problem.

# time parameters
start_time = 0
end_time = 100

# how much time passes between each successive calculation
time_step = 1/4 # in years
end_step = Int((end_time - start_time) / time_step)

# the number of rabbits and foxes when the simulation is started
initial_rabbits = 30000
initial_foxes = 15

# the number of rabbits that must be killed for a fox to be born
rabbits_killed_per_fox_birth = 1000

# the chance a rabbit will die when a rabbit fox cross paths
chance_a_rabbit_will_die_during_a_meeting = 0.5

# the chance that a rabbit and fox will cross paths
chance_of_rabbit_and_fox_meeting = 0.02

# the percent of the rabbit population that will be born this time step
rabbit_growth_rate = 0.2

# the percent of the fox population that will die this time step from old age
fox_death_rate = 0.1

# initialization
rabbits_over_time = fill(0.0, end_step+1)
foxes_over_time = fill(0.0, end_step+1)
model_time = fill(0.0, end_step+1)

rabbits = initial_rabbits
foxes = initial_foxes

rabbits_over_time[1] = rabbits
foxes_over_time[1] = foxes

# simulation
for sim_step = 1:end_step
    # get the time from the step
    sim_time = start_time + sim_step * time_step
    model_time[sim_step] = sim_time
    
    # first we must calculate our flows (our rates)
    rabbits_births = rabbits * rabbit_growth_rate
    rabbits_eaten = min(rabbits, 
      chance_a_rabbit_will_die_during_a_meeting
    * chance_of_rabbit_and_fox_meeting * foxes * rabbits)
    
    fox_births = 1 / rabbits_killed_per_fox_birth * rabbits_eaten
    fox_deathes = foxes * fox_death_rate
    
    # then we update our stocks
    foxes = foxes + fox_births - fox_deathes
    rabbits = rabbits + rabbits_births - rabbits_eaten
    
    # stock values always update in the next time step
    rabbits_over_time[sim_step+1] = rabbits
    foxes_over_time[sim_step+1] = foxes
end

println("Time, Rabbits (thousands), Foxes")
for i = 1:end_step
    print(model_time[i])
    print(",")
    print(rabbits_over_time[i]/1000)
    print(",")
    println(foxes_over_time[i])
end
```
