```julia
countries = ("Japan", "Korea", "China")
```

```julia
for (country, city) in zip(countries, cities)
    println("The capital of $country is $city")
end
```

```julia
for (i, country) in enumerate(countries)
    city = cities[i]
    println("The capital of $country is $city")
end
```
